```python
print('Mayor que:', 2 > 3)#FALSE
print('Menor que:', 2 < 3)#TRUE
print('Mayor o igual que:',3 >= 3)#TRUE
print('Menor o igual que:', 3 <= 4)#TRUE
print('Menor o igual que:', 4 <= 4)#TRUE
```

```python
#operaradores de comparacion de texto

print('1 is 1', 1 is 1) #VERDADERO
print('1 is not 2', 1 is not 2) #TRUE
print('A in Alejandro', 'A' in 'Alejandro') #TRUE
print('L in Alejandro', 'L' in 'Alejandro') #FALSE
print('coding' in 'coding for all') #TRUE
print('a in an:', 'a' in 'an') #TRUE
print('4 is 2 ** 2:', 4 is 2 ** 2) #TRUE
```
