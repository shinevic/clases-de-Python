**AGREGAR COMPONENTES A UNA LISTA**

```python
frutas = ['banana', 'naranja', 'mango', 'limon']
print(frutas)
frutas.append('manzana')
print(frutas)
frutas.append('aguacate')
print(frutas)
```

**INSERTAR COMPONENTES A UNA LISTA**

```python
frutas = ['banana', 'naranja', 'mango', 'limon']
print(frutas)
frutas.insert(2,'manzana')
print(frutas)
frutas.insert(3,'aguacate')
print(frutas)
```

**REMOVER COMPONENTES A UNA LISTA**

```python
frutas = ['banana', 'naranja', 'mango', 'limon', 'manzana', 'aguacate', 'manzana', 'Manzana']
print(frutas)
frutas.remove('manzana')
print(frutas)
frutas.remove('aguacate')
print(frutas)
frutas.remove(frutas[0])
print(frutas)
```
