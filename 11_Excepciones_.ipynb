```python
#sintaxis
try: 
  run this code
except:
  execute this code when there is an error 
else:
  without errors? Run this code.
finally:
  Always run this code.
```

**EJEMPLO DEL ERROR**

```python
numberOne = 5
numberTwo = 1
numberTwo = '1'
```

```python
print(numberOne + numberTwo)
```

**APLICANDO TRY - EXCEPT**

```python
#try - execpt
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except:
  #se ejecuta si se produce una excepcion
  print('Se ha producido un error')
```

**APLICANDO TRY - EXCEPT -ELSE**

```python
numberOne = 1
numberTwo = 2

try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except:
  print('Se ha produciodo un error')
else:
  #se ejecuta si no se produce ninguna excepcion
  print('La ejecucion continua correctamete')
```

```python
numberOne = 1 
numberTwo = 2

try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except:
  print('Se ha producido un error')
else:
  print('La ejecucion continua correctamente')
finally:
  print('La ejecucion sigue')
```

**ANALIZANDO TIPOS DE ERRORES**

```python
numberOne = 5
numberTwo = 1
numberTwo = '1'

print(numberOne + numberTwo)
```

```python
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except TypeError:
     print('se ha producido un error')
```

```python
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except TypeError:
   print('Hubo un error')
   print('Pero ya lo arregle')
   fixone = int(numberOne)
   fixtwo = int(numberTwo)
   print(fixOne + fixTwo)
```

```python
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except ValueError:
   print('Hubo un error')
```

```python
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except TypeError:
  print('Se ha producido un TypeError')
except ValueError:
    print('Hubo un error de valor')
```

**CAPTURA DE TIPO DE EXCEPCION**

```python
numberOne = 5
numberTwo = 1
numberTwo = '1'

try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except TypeError as e:
  print(e)
```
