**IMPORTANDO UN MODULO**

```python
import math
print(math.pi)
print(math.sqrt(16))
print(math.pow(2,3))
print(math.floor(9.84))
```

```python
from math import pi, sqrt, pow
print(pi)
print(sqrt(16))
print(pow(2,3))
print(math.floor(9.84))
```

```python
from math import * 
print(pi)
print(sqrt(16))
print(pow(2,3))
print(floor(9.84))
```
